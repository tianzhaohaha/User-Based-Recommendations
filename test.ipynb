```python

def pearson(X,Y):
    up = 0
    down_x = 0
    down_y = 0
    for i in range(len(X)):
        up+=X[i]*Y[i]
        down_x+=X[i]*X[i]
        down_y+=Y[i]*Y[i]
    return up/(math.sqrt(down_x)*math.sqrt(down_y))
    
```
